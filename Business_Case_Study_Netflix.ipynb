### Type column Analysis with Countplot

### Duration Box plot Analysis

```python
df = data[data['type']=='Movie']
df['duration'] = df['duration'].apply(lambda x: x.split()[0]).astype(int)
sns.boxplot(x = df['duration'])
plt.title("Distribution of duration of movies")
plt.show()
```

```python
df = data[data['type']=='TV Show']
df['duration'] = df['duration'].apply(lambda x: x.split()[0]).astype(int)
sns.boxplot(x = df['duration'])
plt.title("Distribution of duration of movies")
plt.show()
```

#### Observations
 - Most of the movies duration is between 100 and 120
 - Most of the TV shows have seasons between 1 and 2 Seasons
 - Dense Outliers are found in movies when compared TV shows which indicates that there are no fixed timing range for movies and each are varied by few minutes

### How many movies and TV shows were added to Netflix in the years

```python
### Since we unlisted many columns such as Cast, Country, Director, there will be duplicate show_id so group by show_id and then to analysis
df = data.groupby(['show_id','date_added_year','type']).agg(count=('show_id', 'count')).reset_index()
df  = df.groupby(['date_added_year','type']).agg(count=('show_id', 'count')).reset_index()
```

```python
plt.figure(figsize = (10,5))
sns.lineplot(data = df, x = 'date_added_year' , y = 'count' , hue = 'type', marker = 'o' )
plt.xlabel('year_added' , fontsize = 12)
plt.ylabel('total movies/TV shows' , fontsize = 12)
plt.title('total movies and TV shows by the year_added' , fontsize = 12)
plt.show()
```

### How many movies and TV shows that are added to Netflix with respective to release dates

```python
### Since we unlisted many columns such as Cast, Country, Director, there will be duplicate show_id so group by show_id and then to analysis
df = data.groupby(['show_id','release_year','type']).agg(count=('show_id', 'count')).reset_index()
df  = df.groupby(['release_year','type']).agg(count=('show_id', 'count')).reset_index()
```

```python
plt.figure(figsize = (10,5))
sns.lineplot(data = df, x = 'release_year' , y = 'count' , hue = 'type', marker = 'o' )
plt.xlabel('released_year' , fontsize = 12)
plt.ylabel('total movies/TV shows' , fontsize = 12)
plt.title('total movies and TV shows added with respective to thier release date' , fontsize = 12)
plt.show()
```

#### Analysis from above two graphs of released_year and added_year of the TV shows and Movie added to Netflix Collection
 -  We could see movies and TV shows added to the Netflix collectionn peaked during 2019 and 2020.
 -  We could see there is a decline of the movies and TV Shows to Netflix collection most possibily because of **Covid -19** after which movie production and movie release was impacted

```python
df = data[data['country']!='Unkown country']
df = df.groupby(['show_id','country']).agg(count=('country','count')).reset_index()
df.drop('count', axis =1,inplace =True)
top10_country = df['country'].value_counts().iloc[0:10].index
top10_country
```

```python
df = data[data['country']!='Unkown country']
df = df.groupby(['show_id','country','date_added_year']).agg(count=('show_id','count')).reset_index()
df.drop('count', axis =1,inplace =True)
df = df.groupby(['date_added_year','country'])['show_id'].count().reset_index()
df = df[df['country'].isin(top10_country)]
plt.figure(figsize = (15,5))
sns.lineplot(x = df['date_added_year'], y= df['show_id'], hue = df['country']) 
plt.ylabel('Movie/TV Show count')
plt.xlabel('Added_Year')
plt.show()
```

 - Intrestly in **INDIA** Before the start of covid that is by 2018 itself we could see a drop in graph which is further more decreased by covid

### How much Movie and TV Show genres are there in each country 

```python
df = data[(data['country']!='Unkown country')]
df = df[['show_id','country','listed_in']]
df_pivot = df.pivot_table(index = 'country', columns = 'listed_in', aggfunc='size')
plt.figure(figsize=(30, 30))
sns.heatmap(df_pivot, cmap="YlGnBu", linewidths=0.1, linecolor='lightgrey' , cbar_kws={'label': 'Number of Shows'})
plt.title('Number of Movie/ TV shows Genres by Country', fontsize=20)
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Country', fontsize=14)
plt.show()
```

```python
df = data[(data['country']!='Unkown country')& (df['country'].isin(top10_country))]
df = df[['show_id','country','listed_in']]
df_pivot = df.pivot_table(index = 'country', columns = 'listed_in', aggfunc='size')
plt.figure(figsize=(20, 10))
sns.heatmap(df_pivot, cmap="YlGnBu", linewidths=0.1, linecolor='lightgrey' , cbar_kws={'label': 'Number of Shows'})
plt.title('Number of Movie/ TV shows Genres by Top 10 Country', fontsize=20)
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Country', fontsize=14)
plt.show()
```

##### Observations
 - From the above heatmap we can see that no other country other than USA has all the Genres
 - USA most of TV shows and Movies are in Comedies and Dramas
 - India highest TV shows and Movies in International TV Shows
